```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Superstore Level 2.csv")

df.head()
```

```python
df.isnull().sum()
sns.heatmap(df.isnull(), cbar=True)
plt.title("Missing Values Heatmap")
plt.show()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
df[df.duplicated()]
```

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
df['Profit'] = df['Profit'].fillna(df['Profit'].mean())
df.isnull().sum()
```

```python
top_states = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(5)
top_states
```

```python
category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
category_profit
```

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
region_sales
```

```python
df['Month'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum()
monthly_sales
```

```python
sub_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)
sub_profit.head()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Sales', data=df)
plt.title("Sales by Category")
plt.show()
```

```python
monthly_sales.plot(figsize=(10,5))
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
region_sales.plot(kind='bar', figsize=(8,5))
plt.title("Region Wise Sales")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sub_profit.head(10).plot(kind='bar')
plt.title("Top Profitable Sub-Categories")
plt.xlabel("Sub Category")
plt.ylabel("Profit")
plt.show()
```

```python
plt.figure(figsize=(10,6))
scatter_df = df[['Sales','Profit','Category']].dropna()
sns.scatterplot(data=scatter_df,
                x='Sales',
                y='Profit',
                hue='Category')
plt.title("Sales vs Profit by Category")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

```python
plt.figure(figsize=(10,6))
scatter_df = df[['Discount','Profit']].dropna()
sns.scatterplot(data=scatter_df,
                x='Discount',
                y='Profit')
plt.title("Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.show()
```
